```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
fig, ax = plt.subplots(figsize=(4, 3))
ax.plot(x, np.sin(x))
ax.set_title('sin(x)')
```
